```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("sport_athletes.csv")
df
```

```python
#.1 ◊¢◊ë◊ï◊® ◊õ◊ú ◊¢◊†◊£ ◊°◊§◊ï◊®◊ò- ◊î◊¶◊í ◊ê◊™ ◊î◊û◊©◊ß◊ú ◊î◊†◊û◊ï◊ö ◊ë◊ô◊ï◊™◊® , ◊î◊í◊ï◊ë◊î ◊î◊û◊ß◊°◊ô◊û◊ú◊ô ◊ï◊û◊°' ◊î◊ñ◊õ◊ô◊ï◊™ ◊î◊û◊ß◊°◊ô◊û◊ú◊ô ◊ï◊™
result = df.groupby('Sport Type').agg({
    'Weight(kg)': 'min',
    'Height(cm)': 'max',
    'Wins': 'max'
}).reset_index()
print(result)
```

```python
#a. ◊©◊†◊î ◊ê◊™ ◊©◊ù ◊¢◊û◊ï◊ì◊î weight ◊ú- weight_max
#b. ◊©◊†◊î ◊ê◊™ ◊©◊ù ◊¢◊û◊ï◊ì◊î height ◊ú - height_max
#c. ◊©◊†◊î ◊ê◊™ ◊©◊ù ◊î◊¢◊û◊ï◊ì◊î wins ◊ú - wins_max
result = result.rename(columns={
    'Weight(kg)': 'max_weight',
    'Height(cm)': 'max_height',
    'Wins': 'max_wins'
})
print(result)
```

```python
#.3 ◊¶◊ï◊® groups ◊ú◊§◊ô ◊û◊ì◊ô◊†◊î + ◊¢◊†◊£ ◊°◊§◊ï◊®◊ò
#a. ◊î◊¶◊í ◊ê◊™ ◊î◊°◊ò◊ò◊ô◊°◊ò◊ô◊ß◊ï◊™ ◊î◊í◊ë◊ï◊î◊ï◊™ ◊ë◊ô◊ï◊™◊®
df.set_index(['Country', 'Sport Type'], inplace=True)
group_stats = df.groupby(['Country', 'Sport Type']).agg({
    'Height(cm)': 'max',
    'Weight(kg)': 'max',
    'Wins': 'max'
}).reset_index()
group_stats
```

```python
#b. ◊î◊¶◊í ◊ê◊™ ◊î◊°◊ò◊ò◊ô◊°◊ò◊ô◊ß◊ï◊™ ◊î◊í◊ë◊ï◊î◊ï◊™ ◊ë◊ô◊ï◊™◊® ◊®◊ß ◊¢◊ë◊ï◊® ◊õ◊ì◊ï◊®◊°◊ú ◊ë◊ê◊®◊î"◊ë.◊®◊û◊ñ loc
usa_basketball = df.loc[('USA', 'Basketball')]
usa_basketball_stats = usa_basketball[['Height(cm)', 'Weight(kg)', 'Wins']].max()
usa_basketball_stats
```

```python
#c. ◊î◊¶◊í ◊ê◊™ ◊î◊°◊ò◊ò◊ô◊°◊ò◊ô◊ß◊ï◊™ ◊î◊í◊ë◊ï◊î◊ï◊™ ◊ë◊ô◊ï◊™◊® ◊¢◊ë◊ï◊® ◊õ◊ì◊ï◊®◊í◊ú ◊ë◊ú◊ë◊ì. ◊®◊û◊ñ xs
soccer = df.xs('Soccer', level='Sport Type')
soccer_stats = soccer[['Height(cm)', 'Weight(kg)', 'Wins']].max()
soccer_stats
```

```python
#d. ◊î◊¶◊í ◊ê◊™ ◊î◊°◊ò◊ò◊ô◊°◊ò◊ô◊ß◊ï◊™ ◊î◊í◊ë◊ï◊î◊ï◊™ ◊ë◊ô◊ï◊™◊® ◊¢◊ë◊ï◊® ◊ò◊†◊ô◊° ◊ï◊õ◊ì◊ï◊®◊í◊ú ◊ë◊ú◊ë◊ì. ◊®◊û◊ñ isin
selected_sports = df[df.index.get_level_values('Sport Type').isin(['Tennis', 'Soccer'])]
selected_sports_stats = selected_sports.groupby('Sport Type').agg({
    'Height(cm)': 'max',
    'Weight(kg)': 'max',
    'Wins': 'max'
}).reset_index()
selected_sports_stats
```

```python
#e. ◊î◊§◊ï◊ö ◊ê◊™ ◊î◊ê◊ô◊†◊ì◊ß◊°◊ô◊ù ◊ú◊§◊ô ◊°◊§◊ï◊®◊ò-◊û◊ì◊ô◊†◊î ◊ï◊û◊ô◊ô◊ü ◊ê◊™ ◊î◊û◊ô◊ì◊¢ ◊ú◊§◊ô ◊°◊§◊ï◊®◊ò+◊û◊ì◊ô◊†◊î ◊®◊û◊ñ swaplevel
df_swapped = df.swaplevel('Country', 'Sport Type')
df_sorted_swapped = df_swapped.sort_index(level=['Sport Type', 'Country'])
df_sorted_swapped.head()
```

```python
#f. ◊î◊¶◊í ◊ê◊™ ◊î◊°◊ò◊ò◊ô◊°◊ò◊ô◊ß◊ï◊™ ◊© ◊ú ◊¢◊®◊ö: ◊û◊ô◊†◊ô◊û◊ô◊ú◊ô, ◊û◊ß◊°◊ô◊û◊ú◊ô, ◊û◊û◊ï◊¶◊¢ ◊ï◊°◊ò◊ô◊ô◊™ ◊™◊ß◊ü ,
#◊¢◊ë◊ï◊® ◊î◊¢◊û◊ï◊ì◊ï◊™ : ◊û◊°◊§◊® ◊ñ◊õ◊ô◊ï◊™ ◊ï◊í◊ï◊ë◊î. ◊®◊û◊ñ agg
agg_functions = ['min', 'max', 'mean', 'std']
stats = df[['Wins', 'Height(cm)']].agg(agg_functions)
stats
```

```python
#.4 ◊î◊ï◊°◊£ ◊ê◊™ ◊î◊©◊ï◊®◊ï◊™ ◊î◊ë◊ê◊ï◊™ ◊ú◊ò◊ë◊ú◊î ◊ë◊ê◊û◊¶◊¢◊ï◊™ concat:
new_rows = pd.DataFrame({
    'Country': ['Canada', 'USA'],
    'Sport Type': ['Hockey', 'Skiing'],
    'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'],
    'Height(cm)': [183, 172],
    'Weight(kg)': [84, 68],
    'Wins': [3, 4],
    'Olympic Date': ['01/02/2002', '12/02/2010']
})
df_reset = df.reset_index()
df_combined = pd.concat([df_reset, new_rows], ignore_index=True)
df_combined.tail()
```

```python
#- ◊ê◊ó◊ì ◊ê◊™ ◊î◊ò◊ë◊ú◊ê◊ï◊™ ◊©◊ô◊ï◊§◊ô◊¢◊ï ◊®◊ß ◊°◊§◊ï◊®◊ò◊ê◊ô◊ù ◊©◊ô◊© ◊™◊ô◊¢◊ï◊ì ◊ê◊ù ◊î◊ù ◊û◊¢◊©◊†◊ô◊ù ◊ê◊ï ◊ú◊ê . ◊®◊û◊ñ: inner
smoking_url = 'https://raw.githubusercontent.com/pythonai250824/02.04.2025/main/athletes_smoking.csv'
smoking_df = pd.read_csv(smoking_url)
```

```python
#- ◊ê◊ó◊ì ◊ê◊™ ◊î◊ò◊ë◊ú◊ê◊ï◊™ ◊©◊ô◊ï◊§◊ô◊¢◊ï ◊®◊ß ◊°◊§◊ï◊®◊ò◊ê◊ô◊ù ◊©◊ô◊© ◊™◊ô◊¢◊ï◊ì ◊ê◊ù ◊î◊ù ◊û◊¢◊©◊†◊ô◊ù ◊ê◊ï ◊ú◊ê . ◊®◊û◊ñ: inner
merged_inner = pd.merge(df_combined, smoking_df, on='Athlete Name', how='inner')
merged_inner.head()
```

```python
#- ◊ê◊ó◊ì ◊ê◊™ ◊î◊ò◊ë◊ú◊ê◊ï◊™ ◊õ◊ö ◊©◊ô◊ï◊§◊ô◊¢◊ï ◊õ◊ú ◊î◊°◊§◊ï◊®◊ò◊ê◊ô◊ù ◊ï◊™◊ô◊¢◊ï◊ì ◊ê◊ù ◊î◊ù ◊û◊¢◊©◊†◊ô◊ù ◊ê◊ï ◊ú◊ê,
#◊î◊ô◊õ◊ü ◊©◊ê◊ô◊ü ◊™◊ô◊¢◊ï◊ì -- ◊©◊ô◊ï◊§◊ô◊¢ NaN. ◊®◊û◊ñ: outer
merged_outer = pd.merge(df_combined, smoking_df, on='Athlete Name', how='outer')
merged_outer.head()
```
