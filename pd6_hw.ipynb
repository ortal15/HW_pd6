```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("sport_athletes.csv")
df
```

```python
#.1 עבור כל ענף ספורט- הצג את המשקל הנמוך ביותר , הגובה המקסימלי ומס' הזכיות המקסימלי ות
result = df.groupby('Sport Type').agg({
    'Weight(kg)': 'min',
    'Height(cm)': 'max',
    'Wins': 'max'
}).reset_index()
print(result)
```

```python
#a. שנה את שם עמודה weight ל- weight_max
#b. שנה את שם עמודה height ל - height_max
#c. שנה את שם העמודה wins ל - wins_max
result = result.rename(columns={
    'Weight(kg)': 'max_weight',
    'Height(cm)': 'max_height',
    'Wins': 'max_wins'
})
print(result)
```

```python
#.3 צור groups לפי מדינה + ענף ספורט
#a. הצג את הסטטיסטיקות הגבוהות ביותר
df.set_index(['Country', 'Sport Type'], inplace=True)
group_stats = df.groupby(['Country', 'Sport Type']).agg({
    'Height(cm)': 'max',
    'Weight(kg)': 'max',
    'Wins': 'max'
}).reset_index()
group_stats
```

```python
#b. הצג את הסטטיסטיקות הגבוהות ביותר רק עבור כדורסל בארה"ב.רמז loc
usa_basketball = df.loc[('USA', 'Basketball')]
usa_basketball_stats = usa_basketball[['Height(cm)', 'Weight(kg)', 'Wins']].max()
usa_basketball_stats
```

```python
#c. הצג את הסטטיסטיקות הגבוהות ביותר עבור כדורגל בלבד. רמז xs
soccer = df.xs('Soccer', level='Sport Type')
soccer_stats = soccer[['Height(cm)', 'Weight(kg)', 'Wins']].max()
soccer_stats
```

```python
#d. הצג את הסטטיסטיקות הגבוהות ביותר עבור טניס וכדורגל בלבד. רמז isin
selected_sports = df[df.index.get_level_values('Sport Type').isin(['Tennis', 'Soccer'])]
selected_sports_stats = selected_sports.groupby('Sport Type').agg({
    'Height(cm)': 'max',
    'Weight(kg)': 'max',
    'Wins': 'max'
}).reset_index()
selected_sports_stats
```

```python
#e. הפוך את האינדקסים לפי ספורט-מדינה ומיין את המידע לפי ספורט+מדינה רמז swaplevel
df_swapped = df.swaplevel('Country', 'Sport Type')
df_sorted_swapped = df_swapped.sort_index(level=['Sport Type', 'Country'])
df_sorted_swapped.head()
```

```python
#f. הצג את הסטטיסטיקות ש ל ערך: מינימילי, מקסימלי, ממוצע וסטיית תקן ,
#עבור העמודות : מספר זכיות וגובה. רמז agg
agg_functions = ['min', 'max', 'mean', 'std']
stats = df[['Wins', 'Height(cm)']].agg(agg_functions)
stats
```

```python
#.4 הוסף את השורות הבאות לטבלה באמצעות concat:
new_rows = pd.DataFrame({
    'Country': ['Canada', 'USA'],
    'Sport Type': ['Hockey', 'Skiing'],
    'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'],
    'Height(cm)': [183, 172],
    'Weight(kg)': [84, 68],
    'Wins': [3, 4],
    'Olympic Date': ['01/02/2002', '12/02/2010']
})
df_reset = df.reset_index()
df_combined = pd.concat([df_reset, new_rows], ignore_index=True)
df_combined.tail()
```

```python
#- אחד את הטבלאות שיופיעו רק ספורטאים שיש תיעוד אם הם מעשנים או לא . רמז: inner
smoking_url = 'https://raw.githubusercontent.com/pythonai250824/02.04.2025/main/athletes_smoking.csv'
smoking_df = pd.read_csv(smoking_url)
```

```python
#- אחד את הטבלאות שיופיעו רק ספורטאים שיש תיעוד אם הם מעשנים או לא . רמז: inner
merged_inner = pd.merge(df_combined, smoking_df, on='Athlete Name', how='inner')
merged_inner.head()
```

```python
#- אחד את הטבלאות כך שיופיעו כל הספורטאים ותיעוד אם הם מעשנים או לא,
#היכן שאין תיעוד -- שיופיע NaN. רמז: outer
merged_outer = pd.merge(df_combined, smoking_df, on='Athlete Name', how='outer')
merged_outer.head()
```
